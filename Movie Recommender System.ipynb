```python
movies_rating_df = pd.read_csv('u.data', sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
```

```python
movies_rating_df
```

```python
movies_rating_df.drop(['timestamp'],axis=1,inplace=True)
```

```python
movies_rating_df
```

```python
movies_rating_df.describe()
```

```python
movies_rating_df = pd.merge(movies_rating_df, movie_title_df, on = 'item_id')
```

```python
movies_rating_df
```

```python
movies_rating_df.shape
```

```python
movies_rating_df.groupby('title').describe()
```

```python
movies_rating_df.groupby('title')['rating'].describe()
movies_rating_df.groupby('title')['rating'].describe()
```

```python
ratings_df_mean=movies_rating_df.groupby('title')['rating'].describe()['mean']
```

```python
ratings_df_mean
```

```python
movies_count_df=movies_rating_df.groupby('title')['rating'].describe()['count']
```

```python
movies_count_df
```

```python
ratings_mean_count_df=pd.concat([movies_count_df,ratings_df_mean],axis=1)
```

```python
ratings_mean_count_df
```

```python
ratings_mean_count_df.reset_index()
```

```python
ratings_mean_count_df['mean'].plot(bins=100,kind='hist',color='r')
```

```python
ratings_mean_count_df[ratings_mean_count_df['mean']==5]
```

```python
ratings_mean_count_df.sort_values('count',ascending=True).head(100)
```

```python
userid_movietitle_matrix=movies_rating_df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
userid_movietitle_matrix
```

```python
titanic=userid_movietitle_matrix['Titanic (1997)']
titanic
```

```python
titanic_correlations=pd.DataFrame(userid_movietitle_matrix.corrwith(titanic),columns=['Correlations'])
```

```python
titanic_correltaions=titanic_correlations.join(ratings_mean_count_df['count'])
```

```python
titanic_correltaions
```

```python
titanic_correlations.dropna(inplace=True)
titanic_correlations
```

```python
titanic_correlations.sort_values('Correlations',ascending=False)
```

```python
titanic_correlations[titanic_correlations['count']>80].sort_values('Correlations',ascending=False).head(10)
```
